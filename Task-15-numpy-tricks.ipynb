###`Q-1:` Create a random 3x4 matrix with value between 0-100. And perform below tasks

    i. Sort this matrix. np.sort()
    ii. Sort this matrix based on values in 2nd column.
    iii. Sort this matrix based on max value in each row.
    iv. Sort based on elements value.


See examples:
```
arr =
    [[92 90 74]
    [ 6 63 93]
    [15 93 96]
    [70 60 48]]
```
```
i. np.sort
[[74 90 92]
 [ 6 63 93]
 [15 93 96]
 [48 60 70]]

ii. based on 2nd column
[[70 60 48]
 [ 6 63 93]
 [92 90 74]
 [15 93 96]]

iii. based on row max- ascending
[[15 93 96]
 [ 6 63 93]
 [92 90 74]
 [70 60 48]]

iv. based on elements value
[[ 6 15 48]
 [60 63 70]
 [74 90 92]
 [93 93 96]]

```

````python
import numpy as np
n = np.random.randint(1,100,(4,3))
n
````

````python
# sort
np.sort(n)
````

````python
#Based on 2nd column
# n[np.argsort(n[:,1])]
print(n[n[:,1].argsort()])
````

````python
# Sort this matrix based on max value in each row.
print(np.array(sorted(n,key = lambda x:max(x))))
````

````python
# sorted by value
print(np.sort(n,axis= None).reshape(4,3))
````

###`Q-2:` There is an array of marks of 5 students in 4 subjects. Further you are asked to perform below task.
    i. Add marks every student of an extra subject in the same array.
    ii. Add two new students marks in respective 5 subjects.(one subject added in above task)
    iii. Add extra column with sum of all subjects(5-subjects) marks
    iv. Sort the array(non-ascending order) on total marks column--one added in above task. Show top 2 rows.

Note: Change dimension of arrays during concatenation or appending if required.

Given Array-
```
marks = [[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]]

extra_subject = [41, 87, 72, 36, 92]
#Two extra students record-
rec1 = [77, 83, 98, 95, 89]
rec2 = [92, 71, 52, 61, 53]
```

````python
marks = np.array([[13, 10,  9, 33],
       [63, 46, 90, 42],
       [39, 76, 13, 29],
       [82,  9, 29, 78],
       [67, 61, 59, 36]])
extra_subject = np.array([41, 87, 72, 36, 92])
````

````python
#1.
extra = np.expand_dims(extra_subject,axis = 1)
marks = np.concatenate((marks,extra),axis = 1)
print(marks)
````

````python
#2
rec1 = np.array([77, 83, 98, 95, 89]).reshape(1,-1)
rec2 = np.array([92, 71, 52, 61, 53]).reshape(1,-1)
# print(rec1)
# print(rec2)
updated_marks = np.concatenate((marks,rec1,rec2),axis = 0)
print(updated_marks)
````

````python
#3
# sum = np.cumsum(updated_marks,axis = 1).reshape(7,5)[:,-1].reshape(-1,1)
# print(sum)
sum = np.sum(updated_marks,axis = 1,keepdims=True)
# print(sum)
total = np.concatenate((updated_marks,sum),axis = 1)
total
````

````python
#4
# n = (total[total[:,-1].argsort()[::-1]])
# print(n[:2])

print(np.array(sorted(total,key = lambda x: x[-1],reverse = True)[:2]))
````

### `Q-7:` You are given a array. You have to find the minimum and maximum array element and remove that from the array.

```python
import numpy as np

np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))
```

````python
import numpy as np

np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))
# print(np.sort(arr))
# print(np.sort(arr)[0,1:-1])4

max = np.argmax(arr)
min = np.argmin(arr)
print(arr.max())
print(arr.min())
arr = np.delete(arr,max,axis = 1)
arr = np.delete(arr,min,axis = 1)
print(arr)

print(arr.max(),arr.min())
````

### `Q-8:` You are given an arrays. You have to limit this array's elements between 100 to 200. $arr ∈ [100, 700]$. So replace those values accordingly with the minimum and maximum value. Then sort the array and perform the cumulative sum of that array.

````python
arr = np.random.randint(1,200,100)
print(np.cumsum(np.sort(np.clip(arr,a_min=100,a_max=200))))
````

### `Q-9:` You are given a array ($arr ∈ [0, 1]$). First you have round off the elements upto 3 decimal places and compare that
- 0th percentile == minimum value of the array
- 100th percentile == maximum value of the array
- also find the difference betwen 51th percenile and 50th percentile values

````python
a = np.random.random(100)
a = np.around(a,decimals=3)
a
````

````python
print(f"0th percentile == minimum value of the array is {np.percentile(a,0) == a.min()}")
print(f"100th percentile == minimum value of the array is {np.percentile(a,100) == a.max()}")
print(f"The difference betwen 51th percenile and 50th percentile values {np.percentile(a,51) - np.percentile(a,50)}")
````
