###`Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
def change(arr,dtype):
  return pd.to_numeric(arr,downcast = dtype,errors = "coerce")

series = pd.Series([1,2,'Python', 2.0, True, 100])
print(change(series,"float"))
````

````python
d1 = pd.read_csv("/content/items.csv",index_col = ["item_name"])
data = d1.squeeze()
type(data)
````

````python
# print(data[data.isnull()])
data[data.isnull()].size
data.isna().sum()
````

````python
def func(x):
  try:
    y = x[1:]
  except:
    y = x
  return float(y) * 84.34
rupee = data.apply(func)
rupee
````

````python
d = data.dropna()
r = d.apply(lambda x: float(x[1:]) * 84.34)
r
````

````python
data2 = d.apply(lambda x:float(x[1:]))
data2
````

````python
new_data = rupee.fillna(rupee.mean())
new_data
````
