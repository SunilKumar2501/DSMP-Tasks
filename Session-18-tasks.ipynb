Replace old team name as new name before performing any tasks.
```
Delhi Daredevils ->Delhi Capitals

Rising Pune Supergiants -> Rising Pune Supergiant
```

Note: Team1 represents Home team. Exclude not result matches.


````python
dic = {
    "Delhi Daredevils" : "Delhi Capitals",
    "Kings XI Punjab" : "Punjab Kings",
    "Rising Pune Supergiants" : "Rising Pune Supergiant"
}
ipl.replace(dic.keys(),dic.values(),inplace = True)
````

````python
df = ipl[~ipl["WinningTeam"].isna()]
df.head(3)
````

````python
team = df["Team1"].unique()
lst = []
for t in team:
  played = df[(df["Team1"] == t) | (df["Team2"] == t)].shape[0]
  win = df[df["WinningTeam"]==t].shape[0]
  home_win = round(df[(df["WinningTeam"]== t) & (df["Team1"] == t)].shape[0] / df[(df["Team1"] == t)].shape[0] * 100,2)
  away_win = round(df[(df["WinningTeam"]== t) & (df["Team2"] == t)].shape[0] / df[(df["Team2"] == t)].shape[0] * 100,2)
  lst.append([t,played,round(win/played * 100,2),home_win,away_win])
new_df = pd.DataFrame(lst, columns=["Team", "Total_Match_Played", "Win%", "Home_Win%", "Away_Win%"])
new_df.sort_values("Win%",ascending = False,inplace = True)
````

`Team1Players` and `Team2Players` have all players name. It is not a list of players name instead it is str. So handle it as string.

````python
final = ipl[ipl["MatchNumber"] == "Final"]
````

````python

def playername(l):
    # Clean and split the player names, stripping single quotes
    return pd.Series([x.strip("'") for x in l.lstrip("'[").rstrip("]'").split(", ")])

# Initialize an empty list to collect all players
all_players = []

# Process Team1Players
for playerlst in final["Team1Players"]:
    all_players.extend(playername(playerlst).tolist())

# Process Team2Players
for playerlst2 in final["Team2Players"]:
    all_players.extend(playername(playerlst2).tolist())

# Create a Series from the list of all players
players_series = pd.Series(all_players)

# Get the value counts
player_counts = players_series.value_counts().head(5)
player_counts
````

````python
def match_played(df,name):
  return df[(df["Team1"] == name) | (df["Team2"] == name)].shape[0]

def matches_won(df,name):
  return df[df["WinningTeam"] == name].shape[0]

def no_results(df,name):
  return df[((df["Team1"] == name) | (df["Team2"] == name)) & (df["WinningTeam"].isnull())].shape[0]
def points_table(season):
  df = ipl[ipl["Season"] == season]
  new_dframe = pd.DataFrame()
  new_dframe["Team_Name"] = df["Team1"].unique()
  new_dframe["Matches_Played"] = new_dframe["Team_Name"].apply(lambda x : match_played(df,x))
  new_dframe["Matches_Won"] = new_dframe["Team_Name"].apply(lambda x : matches_won(df,x))
  new_dframe["No_Result"] = new_dframe["Team_Name"].apply(lambda x : no_results(df,x))
  new_dframe["Points"] = new_dframe["Matches_Won"] * 2 + new_dframe["No_Result"]
  new_dframe.sort_values(["Points","Team_Name"],ascending = [False,True],inplace = True)
  new_dframe = new_dframe.set_index("Team_Name")
  return new_dframe
````

###`Q-10:` IPL Point Table cont.
Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.

Teams in Top four will have `SeasonPosition` as:
```
    'Winner' - Team won final
    'Runner' - Team lost Final
    3 - Losing Team in Qualifier2
    4 - Losing Team in Eliminator
```

For changing value of pariticular cell use `df.at[row_index, col_label] = value`

Output of two Top 2 in season 2022. Your result should have all teams.
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points   SeasonPosition

Gujarat Titans	    16	   12	       0	     24         Winner
Rajasthan Royals	  17	   10	       0	     20         Runner

```

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
def SeasonPosition(season):
  df3 = points_table(season)
  df = ipl[ipl["Season"] == season].copy()
  df3["Season_Rank"] = df3["Points"].rank(ascending = False, method = "first").astype("object")
  df["Losing_team"] = df.apply(lambda row: row["Team2"] if row["WinningTeam"] == row["Team1"] else row["Team1"],axis = 1)
  final = df[df["MatchNumber"] == "Final"]
  winning_team = final.WinningTeam.values[0]
  runner = final.Losing_team.values[0]
  df3.at[winning_team ,"SeasonRank"] = "Winner"
  df3.at[runner ,"SeasonRank"] = "RunnerUp"
  q = df[df["MatchNumber"] == "Qualifier 2"]
  e = df[df["MatchNumber"] == "Eliminator"]
  qualifier = q.Losing_team.values[0]
  eliminator = e.Losing_team.values[0]
  df3.at[qualifier,"SeasonRank"] = "Third"
  df3.at[eliminator,"SeasonRank"] = "Fourth"
  return df3
````
