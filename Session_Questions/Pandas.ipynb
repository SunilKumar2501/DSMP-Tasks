##Series

```python
#series by list
country = ["India" ,"Pakistan","USA", "Nepal"]
print(pd.Series(country))
```

```python
#integer series
runs = [13,24,56,88,34,100]
run = pd.Series(runs)
```

```python
#custom index
marks = [67,57,89,100]
sub = ["M","S","H","E"]
pd.Series(marks,index = sub)
```

```python
#setting a name
marks = pd.Series(marks,index = sub,name = "Sunil k marks")
marks
```

```python
#using dict
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':100
}

marks_series = pd.Series(marks,name = "Sunil's Marks")
marks_series
```
