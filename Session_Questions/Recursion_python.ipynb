*Recursion is a method to call the function inside the same function.*

 Find the base condition first then look for logic.

```python
# Mutliply
def mul(a,b):
  if b == 1:
    return a
  else:
    return a + mul(a,b-1)
mul(4,8)
```

```python
# Factorial
def fact(number):
  if number == 1:
    return 1
  else:
    return number * fact(number-1)
fact(5)
```

```python
# palandrome
def pal (s):
  if (len(s)) <= 1:
    print("Palandrome")
  else:
    if s[0] == s[-1]:
      pal(s[1:-1])
    else:
      print("Not a palindrome")

pal("maam")
pal("cococa")
pal("madam")
```

```python
# The Rabbit Problem
# fibonachii series
# Most inefficient code

def fib(num):
  if num == 0 or num == 1:
    return 1
  else:
    return fib(num-1) + fib(num-2)
print(fib(5))
```

```python
# Memoization solution
def memo(m,d):
  if m in d:
    return d[m]
  else:
    d[m] = memo(m-1,d) + memo(m-2,d)
    return d[m]


d = {0:0,1:1}
print(memo(20,d))
# print(d)
```
