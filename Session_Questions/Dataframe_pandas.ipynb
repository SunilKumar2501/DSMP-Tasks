```python
#on a dataframe
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def eucl(row):
  one = row["1st point"]
  two = row["2nd point"]

  return ((two[0] - one[0])**2 + (two[1] - one[1])**2)**0.5
```

```python
points_df["distance"] = points_df.apply(eucl,axis = 1)
```
