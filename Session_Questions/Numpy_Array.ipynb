# Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1.astype(np.int32)
```

```python
# max,sum,min,prod
np.max(a1)
np.sum(a1)
np.min(a1)
np.prod(a1)
```

```python
# these function accourding to the row and columns
# here 0 is column and 1 is row
print(np.max(a1,axis = 0)) #by column
print(np.max(a1,axis = 1)) # by row
print(np.min(a1,axis = 0))
print(np.min(a1,axis = 1))
```

```python
# mean/median/std/var
print(a1)
print(np.mean(a1))
print(np.median(a1))
print(np.std(a1))
print(np.var(a1))
```

```python
# trignometric
np.sin(a1)
```

```python
# Dot product -> first matric col and second matrix row must have same shape.
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)
print(a2)
print(a3)
print(np.dot(a2,a3))
```

```python
# log and exponents
np.log(a1)
```

```python
np.exp(a1)
```

```python
# round/ciel/floor
a4 = np.round(np.random.random((3,4))* 100)
print(a4)

a4 = np.ceil(np.random.random((3,4))* 100)
print(a4)

a4 = np.floor(np.random.random((3,4))* 100)
print(a4)
```
