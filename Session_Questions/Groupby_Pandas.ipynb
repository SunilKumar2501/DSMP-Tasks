```python
#looping on groups
# df = pd.DataFrame(columns = movies.columns)
# for group,data in genre:
#   df = df.append(data[data["IMDB_Rating"] ==data["IMDB_Rating"].max()])
# print(df)

df_list = []

# Loop through each group and data in 'genre'
for group, data in genre:
    # Find the row(s) with the maximum 'IMDB_Rating' in each group
    max_rating_data = data[data["IMDB_Rating"] == data["IMDB_Rating"].max()]
    # print(type(max_rating_data))
    # Append this data to the list
    df_list.append(max_rating_data)

# Concatenate the list into a single DataFrame
df = pd.concat(df_list, ignore_index=True)

# Display the result
df = df.set_index("Series_Title")
df
```

```python
#num of movies starting with A
def foo(group):
  ans = group[group["Series_Title"].str.startswith("A")]
  return ans

genre.apply(foo) #split apply combine
```

```python
#ranking on every movies in each group on imdb score
def rank_movie(group):
  group["genre_rank"] = group["IMDB_Rating"].rank(ascending = False,method = "first")
  return group
genre.apply(rank_movie)
```

```python
#Normalize rating of each movie
```

```python
def normal(group):
  group["norm_rating"] = round((group["IMDB_Rating"] - group['IMDB_Rating'].min()) / (group["IMDB_Rating"].max() - group['IMDB_Rating'].min()),2)
  return group
```

```python
genre.apply(normal)
```

```python
ipl = pd.read_csv("/content/drive/MyDrive/Datasets/Copy of deliveries.csv")
ipl.head()
```

```python
ipl.shape
```

```python
#top 10 batsman in terms of runs scored
batsman = ipl.groupby("batsman")
batsman["batsman_runs"].sum().sort_values(ascending = False).head(10)
```

```python
#batsman with most sixes in ipl
six = ipl[ipl["batsman_runs"] == 6]
six.groupby("batsman")["batsman"].count().sort_values(ascending = False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp = ipl[ipl["over"] > 15]
temp = temp[(temp["batsman_runs"] == 4) | (temp["batsman_runs"] == 6)]
# temp["batsman"].value_counts()
temp.groupby("batsman")["batsman"].count().sort_values(ascending=False).head(10)
```

```python
#find V Kohli's record agains all teams
temp2 = ipl[ipl["batsman"] == "V Kohli"]
temp2.groupby("bowling_team")["batsman_runs"].sum().reset_index()
```

```python
#function that return highest score of any batsman
def highest(batsman):
  temp_df = ipl[ipl["batsman"] == batsman]
  return temp_df.groupby(["bowling_team","match_id"])["batsman_runs"].sum().sort_values(ascending = False).head(5)
```
