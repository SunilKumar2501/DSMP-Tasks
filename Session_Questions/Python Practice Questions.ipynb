```python
# Q1 Given the scores for a season, determine the number of times Maria breaks her records for
# most and least points scored during the season.
def breakingRecords(n,scores):
  result = [0,0]
  t = scores[0]
  a = t
  b = t
  for i in scores:
    if(a > i):
      result[1] = result[1] + 1
      a = i
    if(b < i):
      result[0] = result[0] + 1
      b = i
  return result

scores = [10, 5, 20 ,20, 4, 5, 2, 25, 1]
n = len(scores)
breakingRecords(n,scores)
```

# Problem-2: Student management system in Python.
# Write a program to build a simple Student Management System using Python which can perform the following operations:
1. Accept - This method takes details from the user like name, roll number, and marks
for two different subjects.
2. Display - This method displays the details of every student.
3. Search - This method searches for a particular student from the list of students. This
method will ask the user for roll number and then search according to the roll
number
4. Delete - This method deletes the record of a particular student with a matching roll
number.
5. Update - This method updates the roll number of the student. This method will ask
for the old roll number and new roll number. It will replace the old roll number with
a new roll number

```python
class Student:
    lst = []  # Class variable to store all students

    def __init__(self, name, roll, m1, m2):
        self.name = name
        self.Roll = roll
        self.Marks1 = m1
        self.Marks2 = m2


    def main(self):
        while True:
            print("""
            Choose from the given options:
            1. Display the Students' details
            2. Search for the students
            3. Delete a student's details
            4. Update the roll number
            5. Exit
            """)
            choice = int(input("Enter the option: "))
            if choice == 1:
                self.Display()
            elif choice == 2:
                self.Search()
            elif choice == 3:
                self.Delete()
            elif choice == 4:
                self.Update()
            elif choice == 5:
                print("Thank you")
                break  # Exit the loop to stop the program
            else:
                print("Invalid choice. Please choose again.")

    def Accept(self, name, roll, m1, m2):
        obj = Student(name, roll, m1, m2)
        Student.lst.append(obj)

    def Display(self):
        for i in Student.lst:
            print("Student Name :", i.name)
            print("Roll No. :", i.Roll)
            print("Math marks :", i.Marks1)
            print("English marks :", i.Marks2)
            print("")

    def Search(self):
        roll = int(input("Enter the roll number to search: "))
        for i in Student.lst:
            if i.Roll == roll:
                print("Student found:")
                print("Student Name :", i.name)
                print("Roll No. :", i.Roll)
                print("Math marks :", i.Marks1)
                print("English marks :", i.Marks2)
                return
        print("Student not found.")

    def Delete(self):
        roll = int(input("Enter the roll number to delete: "))
        for i in Student.lst:
            if i.Roll == roll:
                Student.lst.remove(i)
                print(f"Student with Roll No. {roll} has been deleted.")
                return
        print("Student not found.")

    def Update(self):
        roll = int(input("Enter the roll number to update: "))
        for i in Student.lst:
            if i.Roll == roll:
                new_roll = int(input("Enter the new roll number: "))
                i.Roll = new_roll
                print(f"Student's roll number has been updated to {new_roll}.")
                return
        print("Student not found.")

# Create an instance of the Student class and accept some initial students
obj = Student("Sunil", 0, 50, 30)
obj.Accept("A", 1, 100, 100)
obj.Accept("B", 2, 90, 90)
obj.Accept("C", 3, 80, 80)

# Run the main menu
obj.main()
```
