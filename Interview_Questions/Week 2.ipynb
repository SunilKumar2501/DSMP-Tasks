```python
a = {'name':'nitish','age':33}
# aliasing->cloning
b = a.copy()

b['gender'] = 'male'

print(b)
print(a)
```

```python
# SHALLOW CLONING
# [:] -> shallow copy
a = [1,2,3]
# shallow copy
b = a.copy()

b.append(4)

print(a)
print(b)
```
